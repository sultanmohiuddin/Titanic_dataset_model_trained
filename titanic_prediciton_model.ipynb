```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("C://Users//user//Desktop//titanic//train.csv")
```

```python
df.head()
```

# checking columns containing NaN values

```python
df.isnull().sum()
```

# filling out the NaN values, so that these value dont cause issue in prediction model

```python
df.fillna(df.mean(),inplace=True)
```

# model training

```python
feature_col=['PassengerId','Pclass','Age','SibSp','Parch','Fare']
x=df[feature_col].values
y=df.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
result=model.predict(X_test)
result
```
